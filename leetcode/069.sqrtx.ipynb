### [LeetCode]题解（python）：069-Sqrt(x)

### 题目来源：
https://leetcode.com/problems/sqrtx/

### 题意分析：
求一个数的平方根，如果为小数，返回整数部分即可

### 代码python：

```python
'''
思路一：二分法
'''
class Solution(object):
    def mySqrt(self, x):
        """
        :type x: int
        :rtype: int
        """
        lo = 0
        hi = x
        while lo <= hi:
            mid = (lo + hi) /2
            v = mid * mid
            if v < x:
                lo = mid + 1
            elif v > x:
                hi = mid - 1
            else:
                return mid
        return hi
                          
```

```python
'''
思路二：公式 牛顿公式
'''
class Solution(object):
    def mySqrt(self, x):
        """
        :type x: int
        :rtype: int
        """
        r = x
        while r * r > x:
            r = (r + x/r) /2
        return r
```
