### [LeetCode]题解（python）：013-Roman to Integer

### 题目来源：
https://leetcode.com/problems/roman-to-integer/

### 题意分析：
将罗马数字转换为整数

### 解题思路：
#### 规律
1. 相同的数字连写，所表示的数等于这些数字相加得到的数，如 Ⅲ=3；
2. 小的数字在大的数字的右边，所表示的数等于这些数字相加得到的数，如 Ⅷ=8、Ⅻ=12；
3. 小的数字（限于 Ⅰ、X 和 C）在大的数字的左边，所表示的数等于大数减小数得到的数，如 Ⅳ=4、Ⅸ=9；

### 代码python：

```python
'''
先做一个字符和数值对应的字典，{'I':1,'V':5,'X':10,'L':50,'C':100,'D':500,'M':1000}；
从前向后扫描，如果发现输入的字符串后一位小于等于前一位，相加；如果后一位大于前一位，相减。
'''
class Solution(object):
    def romanToInt(self, s):
        """
        :type s: str
        :rtype: int
        """
        d = {"I":1, "V": 5, "X":10,"L":50,"C":100, "D":500, "M":1000}
        ans = 0
        # 循环到字符串最后一位的前一个字符，最后一个字符已没有比较对象
        for i in xrange(0, len(s) - 1):
            cur = s[i]
            curafter = s[i+1]
            if d[cur] < d[curafter]:
                ans -= d[cur]
            else:
                ans += d[cur]
        # 直接加最后一个字符即可
        ans += d[s[-1]]
        return ans
```

1. xrange 用法与 range 完全相同，所不同的是生成的不是一个list对象，而是一个生成器。
2. 要生成很大的数字序列的时候，用xrange会比range性能优很多，因为不需要一上来就开辟一块很大的内存空间，这两个基本上都是在循环的时候用。
3. xrange则不会直接生成一个list，而是每次调用返回其中的一个值，所以xrange做循环的性能比range好，尤其是返回很大的时候。尽量用xrange，除非你是要返回一个列表。
